```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("../data/tmdb_5000_movies.csv")
credits = pd.read_csv("../data/tmdb_5000_credits.csv")

movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on="title")
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]

movies.head()
```

```python
import ast
def Convert(obj):
    L = []
    for name_idx in ast.literal_eval(obj):
        L.append(name_idx['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(Convert)
```

```python
movies['keywords'] = movies['keywords'].apply(Convert)
```

```python
def Cast_name(obj):
    L = []
    for name_idx in ast.literal_eval(obj):
        L.append(name_idx['name'])
    return L[:3]
```

```python
movies['cast'] = movies['cast'].apply(Cast_name)
```

```python
def fetch_director(obj):
    L = []
    
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def Whitespaces(obj):
    L = []
    for element in obj:
        L.append(element.replace(' ',''))
    return L
```

```python
movies['genres'] = movies['genres'].apply(Whitespaces)
movies['cast'] = movies['cast'].apply(Whitespaces)
movies['crew'] = movies['crew'].apply(Whitespaces)
movies['keywords'] = movies['keywords'].apply(Whitespaces)
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
movies['overview'].isnull().sum()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies_df = movies[['movie_id','title','tags']]
```

```python
new_movies_df.head()
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x : [text.lower() for text in x])
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x : ' '.join(x))
```

```python
new_movies_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_movies_df['tags']).toarray()
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    
    movie_index = new_movies_df[new_movies_df['title'] == movie].index[0]
    
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)),
                         reverse=True,
                         key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_movies_df.iloc[i[0]].title)
```

```python
recommend("Spectre")
```

```python
import pickle

pickle.dump(new_movies_df, open('movies.pkl','wb'))
pickle.dump(similarity, open('similarity.pkl','wb'))
```
